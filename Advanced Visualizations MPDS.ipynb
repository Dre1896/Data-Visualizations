### Box Plots
Box plots are good to show the spread of values, median, and outliers of the data
- Q3: This represents the 75th percentile value of the data. It's commonly referred to as the upper hinge.

- Q1: This represents the 25th percentile value of the data. It's commonly referred to as the lower hinge.

- Box: It's commonly referred to as a step. It represents the difference between the upper hinge and the lower hinge.

- Median: This is the midpoint of the data.

- Max: This represents the upper inner fence. It is 1.5 times the step above Q3.

- Min: This represents the lower inner fence. It is 1.5 times the step below Q1.

Values above the Max or lesser than the Min are called outliers, also known as fliers.

```python
## Creating some data
np.random.seed(20)
box_data_1 = np.random.normal(110, 20, 210)
box_data_2 = np.random.normal(70, 20, 190)
box_data_3 = np.random.normal(85, 40, 220)

## combining the different data in a list
data_to_plot = [box_data_1, box_data_2, box_data_3]

# create the boxplot
bp = plt.boxplot(data_to_plot)
```

```python
# Ways to style boxplots
bp = plt.boxplot(data_to_plot, patch_artist = True)
```

```python
## ways to change the outline color or how to fill color and linewidth of boxes
for box in bp['boxes']:
    # change the color of the outline
    box.set( color = '#7570b3', linewidth = 3)
    # change fill color
    box.set( facecolor = '#1b9e77')
```

```python
## changing the color and linewidth of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color = '#7570b3', linewidth = 3)
```

```python
## changing the color and linewidth of the caps
for cap in bp['caps']:
    cap.set(color = '#7570b3', linewidth = 3)
```

```python
## changing the color and linewidth of the medians
for median in bp['medians']:
    median.set(color = '#b2df8a', linewidth = 3)
```

```python
## changing the style of fliers and their fill
for flier in bp['fliers']:
    flier.set(marker = 'o', color = 'e#7298a', alpha = 0.5)
```

### Heatmaps
The heatmap is a classic representation where individuals values of a matrix are shown as colors. They help visualize concentration of values between two dimensions of a matrix. This makes finding patterns easier and gives perspectives of depth.

```python
data = np.random.rand(10, 6)
rows = list('ZYWVUTSRQ') # This will be the Y-Label
columns = list('ABCDEF') # This will be the X-Label

plt.pcolor(data) # manipulates the color
plt.show()
```

### Scatter plots with histograms
It's pretty simple to combine scatter plots with histograms on each ais. These plots help show the distribution of values on each axis.

```python
from matplotlib.ticker import NullFormatter
# the random data
x = np.random.randn(1000)
y = np.random.randn(1000)

nullfmt = NullFormatter()
```

```python
# define the labels for the axes
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
bottom_h = left_h = left + width + 0.03

rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom_h, width, 0.3]
rect_histy = [left_h, bottom, 0.2, height]

# Once the size and height are dfined, we see the axes are plotted for the scatter plot 
# this is also true for the histograms

plt.figure(1, figsize = (8, 8))
axScatter = plt.axes(rect_scatter)
axHistx = plt.axes(rect_histx)
axHisty = plt.axes(rect_histy)

axHistx.xaxis.set_major_formatter(nullfmt)
axHisty.yaxis.set_major_formatter(nullfmt)
axScatter.scatter(x, y)

binwidth = 0.35
xymax = np.max([np.max(np.fabs(x)), np.max(np.fabs(y))])
lim = (int(xymax / binwidth) + 1) * binwidth
axScatter.set_xlim((-lim, lim))
axScatter.set_ylim((-lim, lim))

bins = np.arange(-lim, lim + binwidth, binwidth)

axHistx.hist(x, bins=bins)
axHisty.hist(y, bins = bins, orientation = 'horizontal')
axHistx.set_xlim( axScatter.get_xlim())
axHisty.set_ylim( axScatter.get_ylim())
plt.show()
```

### Bubble Charts
These are essentially scatter plots with an added dimension. The added dimension aids in setting the size of the bubble, meaning that the greater the size of the bubble, the larger the value that represents the bubble. This kind of a chart helps in analyzing the data of three dimensions.

```python
plt.style.use('ggplot')
df = pd.DataFrame(np.random.rand(50, 3), columns = ['a', 'b', 'c'])
df.plot(kind = 'scatter', x = 'a', y = 'b', s = df['c']*400)
```

### Hexagon Bin Plots
These can be created using a dataframe function and adding the modification, kind = 'hexbin'. These plots are helpful when scatter plots become too dense to actually make sense of. This bins the spatial area of the charts and the intensity of the color that heagons can be interpreted as different points that show higher concentration in their respective areas.

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns = ['a', 'b'])
df['b'] = df['b'] + np.arange(1000)
df.plot(kind = 'hexbin', x = 'a', y = 'b', gridsize = 25, color = 'white')
```

### 3D plots of a surface

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = 'hot')
```

Below is the same chart at 0 degrees of elevation.

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.view_init(elev = 0., azim = 0)
ax.plot_surface(X, Y,Z, rstride = 1, cstride = 1, cmap = 'hot')
```

Below is the same chart at 50 degrees of elevation.

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.view_init(elev = 50., azim = 0)
ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = 'hot')
```

This will be the last 3d plot. It is the same plot, at 50 degrees elevation and viewed at a 30 degree angle.

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.view_init(elev = 50., azim = 30)
ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = 'hot')
```
